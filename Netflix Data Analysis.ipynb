```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Dataframe
df = pd.read_csv('mymoviedb.csv', lineterminator = '\n')
```

```python
df.head()
```

```python
#info
print(df.info())
```

```python
#Statistics
df.describe()
```

```python
#Check duplicates
df.duplicated().sum()
```

```python
#Change wrong data types
df['Release_Date'] = pd.to_datetime(df['Release_Date'])
print(df['Release_Date'].dtypes)
```

```python
#Show only year
df['Release_Date'] = df['Release_Date'].dt.year
print(df['Release_Date'].dtypes)
print(df.head())
```

```python
#Remove unwanted columns
col = ['Overview','Original_Language','Poster_Url']
df.drop(col, axis=1, inplace=True)
df.columns
```

```python
df.head()
```

```python
#Categorize data for Vote_Average column
def categorize(df, col, labels):
    edges = [df[col].describe()['min'],df[col].describe()['25%'],df[col].describe()['50%'],
             df[col].describe()['75%'],df[col].describe()['max']]

    df[col] = pd.cut(df[col], edges, labels = labels, duplicates = 'drop')
    return df

labels = ['not_popular','below_avg','average','popular']


categorize(df, 'Vote_Average', labels)

print(df['Vote_Average'].unique())
print(df['Vote_Average'].value_counts())
```

```python
df.head()
```

```python
#check null values
print(df.isna().sum())
```

```python
#delete null values
df.dropna(inplace=True)
```

```python
#recheck null values
print(df.isna().sum())
```

```python
# split genres into list from genre column
df['Genre']= df['Genre'].str.split(', ')
df = df.explode('Genre').reset_index(drop=True)

df.head()
```

```python
#casting column into category

df['Genre']= df['Genre'].astype('category')
df['Genre'].dtypes
```
